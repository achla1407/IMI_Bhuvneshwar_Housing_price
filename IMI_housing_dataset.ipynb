PROBLEMS OBSERVED IN THE DATASET
1) max. Bedrooms 33 , but the area doesn't justify , there must be the error in data entry.
2) Datatypes of some column are wrong !

```python
#   the date datatype
df['date'] = pd.to_datetime(df['date'])
```

```python
# It contains '?' which turns the whole column into strings.
# We replace '?' with NaN, convert to float, and then fill NaNs (usually with 0 if no basement)
df['sqft_basement'] = df['sqft_basement'].replace('?', np.nan)
df['sqft_basement'] = df['sqft_basement'].astype(float)
df['sqft_basement'] = df['sqft_basement'].fillna(0)
```

```python
#outlier
# Locate the row with 33 bedrooms and change it to 3 (based on sqft evidence)
df.loc[df['bedrooms'] == 33, 'bedrooms'] = 3
```

```python
df.describe()
```

DEALING WITH PRICE COLUMN

```python
# coverting price into lakhs for readability and intution.
if 'price' in df.columns:
    df['price'] = df['price'] / 100000
    # Renaming
    df = df.rename(columns={'price': 'price (in lakhs)'})
elif 'price (in lakhs)' in df.columns:
    pass
else:
    print("Error: Expected column 'price' or 'price (in lakhs)' not found in DataFrame.")

if 'price (in lakhs)' in df.columns:
    df['price (in lakhs)']
```

FIXING DATE AND TIMESTAMP

```python
df['date'][0]
# removing time stamp
df['date'] = df['date'].dt.date
```

 Fixing yr (built and renovated )(float to Int & Feature Engineering)

```python
#Datatype and handling NAN
df['yr_built'] = df['yr_built'].fillna(0).astype(int)
df['yr_renovated'] = df['yr_renovated'].fillna(0).astype(int)

# Creating actionable 'Age' feature
# (0 if renovated recently is better than 0 if never renovated, so we prioritize renovation year)
current_year = 2015 # Based on dataset
df['house_age'] = current_year - df['yr_built']
df['years_since_reno'] = df.apply(lambda x: (current_year - x['yr_renovated']) if x['yr_renovated'] != 0 else (current_year - x['yr_built']), axis=1)

```

DEALING WITH AREAS
1) to handle massive mansions that skew the data

```python
#Outlier Working through log transformation

df['log_sqft_living'] = np.log1p(df['sqft_living'])
```

DEALING WITH LOCATION


1) Zipcode , lat and long

```python

plt.figure(figsize=(10, 8))
# Plot all houses, colored by price
sns.scatterplot(data=df, x='long', y='lat', hue='price (in lakhs)', alpha=0.7, palette='coolwarm')

# Mark Seattle City Center with a red star
plt.scatter(-122.3321, 47.6062, color='red', s=100, marker='*', label='Seattle Center')

plt.legend()
plt.title('Seattle Housing Map')
plt.show()

```

```python
# Zipcode is categorical, not numerical. Let's convert it to string so models don't confuse
df['zipcode'] = df['zipcode'].astype(str)

# Calculate distance to Seattle Center (approx Lat: 47.6062, Long: -122.3321)(from the Dataset)
# We use simple Euclidean distance for speed (approximate)
seattle_lat = 47.6062
seattle_long = -122.3321

df['lat_diff'] = df['lat'] - seattle_lat
df['long_diff'] = df['long'] - seattle_long
# Distance formula: sqrt(x^2 + y^2)
df['dist_to_seattle'] = np.sqrt(df['lat_diff']**2 + df['long_diff']**2)

# CALCULATE DISTANCE IN KM
# Using the Haversine Formula for accuracy
def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of Earth in km
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    dphi = np.radians(lat2 - lat1)
    dlambda = np.radians(lon2 - lon1)
    a = np.sin(dphi/2)**2 + np.cos(phi1)*np.cos(phi2)*np.sin(dlambda/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c

# Seattle City Center Coordinates
seattle_lat = 47.6062
seattle_long = -122.3321

df['dist_to_seattle_km'] = haversine_km(df['lat'], df['long'], seattle_lat, seattle_long)
```

DEALING WITH : GRADE , CONDITION , VIEW AND WATERFRONT


1)Combines the best categorical features into one powerful number TO MAKE LUXURY SCORE
2)We give 'waterfront' a x3 multiplier because it's a huge price driver.



```python
df['luxury_score'] = df['grade'] + df['condition'] + df['view'] + (df['waterfront'] * 3)
```

DROP UNNECESSARY COLUMNS

```python

columns_to_drop = ['id', 'lat_diff', 'long_diff']
# Use errors='ignore' to prevent KeyError if a column is already dropped
df = df.drop(columns=columns_to_drop, errors='ignore')

print(df[['dist_to_seattle_km', 'luxury_score', 'price (in lakhs)']].head())

print("\nCorrelation Check (Does the new feature work?):")
print(df[['price (in lakhs)', 'dist_to_seattle_km', 'luxury_score']].corr()['price (in lakhs)'])

# Save the final engineered dataset
df.to_csv('imi_engineered.csv', index=False)
```

## DATA VISUALISATION

```python

sns.set_style("whitegrid")

#  Correlation Heatmap (The Big Picture)
# Helps to see which variables are most strongly connected to Price
plt.figure(figsize=(12, 10))
numeric_df = df.select_dtypes(include=[np.number]) # Only numbers
corr = numeric_df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool)) # Hides the upper duplicate triangle
sns.heatmap(corr, mask=mask, cmap='coolwarm', vmax=1, vmin=-1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title('Correlation Heatmap: What drives the Price?', fontsize=16)
plt.savefig('correlation_heatmap.png')
plt.show()

# Price vs. Distance to Seattle (Location Analysis)
# proves that "Closer = More Expensive"
plt.figure(figsize=(10, 6))
sns.scatterplot(x='dist_to_seattle_km', y='price (in lakhs)', data=df, alpha=0.5, hue='price (in lakhs)', palette='viridis')
# Add a trend line to make the relationship obvious
sns.regplot(x='dist_to_seattle_km', y='price (in lakhs)', data=df, scatter=False, color='red')
plt.title('Impact of Location: Price vs. Distance to Seattle', fontsize=14)
plt.xlabel('Distance to City Center (km)')
plt.ylabel('Price (in Lakhs)')
plt.savefig('price_vs_distance.png')
plt.show()

#  Price vs. Luxury Score (Luxury Analysis)
# Shows how your new feature separates cheap houses from expensive ones
plt.figure(figsize=(10, 6))
sns.boxplot(x='luxury_score', y='price (in lakhs)', data=df, palette='coolwarm')
plt.title('Impact of Quality: Price vs. Luxury Score', fontsize=14)
plt.xlabel('Luxury Score (Grade + View + Condition)')
plt.ylabel('Price (in Lakhs)')
plt.savefig('price_vs_luxury.png')
plt.show()

# Price vs. View (View Analysis)
plt.figure(figsize=(10, 6))
sns.boxplot(x='view', y='price (in lakhs)', data=df, palette='Set2')
plt.title('Impact of View: Price vs. View Rating (0-4)', fontsize=14)
plt.xlabel('View Rating')
plt.ylabel('Price (in Lakhs)')
plt.savefig('price_vs_view.png')
plt.show()

# Geospatial Map (Visual Proof)
# Plots every house on a map, colored by price
plt.figure(figsize=(10, 8))
sns.scatterplot(x='long', y='lat', data=df, alpha=0.4, hue='price (in lakhs)',
                palette='coolwarm', size='price (in lakhs)', sizes=(10, 200))
plt.title('Geospatial Map: Where are the expensive houses?', fontsize=14)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.savefig('geospatial_map.png')
plt.show()
```
